```python
groupby_df = fillna_df.groupby('일시').mean()
```

```python
groupby_df.drop("지점", axis=1).reset_index()
```

```python
from scipy.stats import skew

features_index = groupby_df.dtypes[origin_df.dtypes != 'object'].index

skew_features = groupby_df[features_index].apply(lambda x : skew(x))

skew_features_top = skew_features[skew_features > 1]
print(skew_features.sort_values(ascending=False))
```

```python
groupby_df.describe()
```

 ** 1이상 log 변환**

```python
groupby_df[skew_features_top.index] = np.log1p(groupby_df[skew_features_top.index])
```

```python
skew_features_logs = groupby_df[features_index].apply(lambda x : skew(x))

print(skew_features_logs.sort_values(ascending=False))
```
