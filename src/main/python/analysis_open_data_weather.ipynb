### 합계 일사량은 조사자체를 안하는 지역이 많아서 null value가 많다 ( 삭제 )

** 숫자 피처들에 대한 데이터 분포 왜곡도 확인 후 높은 왜곡도를 가지는 피처 추출 **

```python
from scipy.stats import skew

features_index = clean_df.dtypes[(clean_df.dtypes != 'object') & (clean_df.dtypes != 'datetime64[ns]') ].index

skew_features = clean_df[features_index].apply(lambda x : skew(x))

skew_features_top = skew_features[skew_features > 1]
print(skew_features_top.sort_values(ascending=False))
#print(skew_features.sort_values(ascending=False))
```

### check skew

```python
skew_features.to_frame().rename(columns={0:"skew"})
```

** 왜곡도가 1인 피처들은 로그 변환 적용 **

```python
import numpy as np
origin_df[skew_features_top.index] = np.log1p(origin_df[skew_features_top.index])
```
